```python
#2-1두 개의 정수를 입력받아서 다음과 같이 4가지 관계 연산자의 결과를 출력하시오. 이 때 입력받은 두 정수를 이용하여 출력하시오
#입력값 4 5 , 출력 4>5---0 , 4< 5---1, 4>=5---0 , 4<=5---1

a, b = map(int, input().split())
#if(a<b):
  #print(a,'>', b ,'---',0 ) #왼쪽 값이 오른쪽 값보다 크다
  #print(a,'<', b ,'---',1 ) #왼쪽 값이 오른쪽 값보다 작다
  #print(a,'>=',b ,'---',0 ) #왼쪽 값이 오른쪽 값보다 크거나 동일하다
  #print(a,'<=',b ,'---',1 ) #왼쪽 값이 오른쪽 값보다 작거나 동일하다

  print(a,'>', b ,'---',a > b ) #왼쪽 값이 오른쪽 값보다 크다
  print(a,'<', b ,'---',a < b ) #왼쪽 값이 오른쪽 값보다 작다
  print(a,'>=',b ,'---',a >= b ) #왼쪽 값이 오른쪽 값보다 크거나 동일하다
  print(a,'<=',b ,'---',a <=b) #왼쪽 값이 오른쪽 값보다 작거나 동일하다

  
```

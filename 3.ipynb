```python
#3-1 정수를 입력받아 0이면 zero, 양수면 plus , 음수면 minus
a = int(input(""))

if a == 0:
  print("zero")

elif (a>0):
  print("plus")

else:
  a<0
  print("minus")
```

```python
#3-2 영문 대문자를 입력받아 A,B,C,D,F 이면 출력 그 외 문자이면 error 라고 출력
text = input("")
text1 = (text.upper()) #대문자로 변환
if (text1 == 'A'):
  print("Excellent")
elif (text1 == 'B'): 
  print("Good")
elif (text1 == 'C'):
  print("Usually")
elif (text1 == 'D'):
  print("Effort")
elif (text1 == 'F'): 
  print("Failure")
else: 
  print("error")
```

```python
#3-3 1번은 개 , 2번은 고양이, 3번은 병아리로 정하고 번호를 입력하면 동물을 영어로 출력하는 프로그램

num = int(input("Number?"))

if (num==1):
  print("dog")

elif (num==2):
  print("cat")

elif (num==3):
  print("chick")

else:
  print("I don\'t know.")
```

```python
#3-4 정수 1~12 사이 에서 입력받아 평년의 경우 해당 월 수를 출력
month = int(input("월을 입력하시오:"))
list = [ 1, 3, 5, 7, 8, 10, 12]
list1 = [4, 6, 9, 11]
list2 = [2]
if month in list :
  print(31,"일")

elif month in list1 :
  print(30,"일")
elif month in list2 :
  print(28,"일")
else:
  print("잘못 입력했습니다.")
```

```python
#3-5 남자는 M, 여자는 F 로 나타내고 18세 이상을 성인으로 하자 
#성별을 입력받아 MAN , WOMAN , BOY, GIRL, 로 구분 예 F 15 = GIRL

gender = input("성별을 입력하시오:")
age    = int(input("나이를 입력하시오:"))


if (gender == 'M') and (age>= 18):
  print("MAN")
elif (gender == 'M') and (age > 18):
  print("BOY")
elif (gender == 'F') and (age >= 18):
  print("WOMAN")
else:
  print("GIRL")
```

```python
#3-6 년도를 입력받아 윤년(leap year) 인지 평년(common year)인지 판단하는 프로그램 작성
# 2008 , leap year
#400으로 나누어 떨어지면 윤년 또는 4로 나누어 떨어지고 100으로 나누어 떨어지지 않으면 윤년이다. 나머지는 평년이다


year = int(input(''))
if (((year%4 == 0) and (year% 100 !=0)) or year % 400 == 0):
  print("leap year")

else:
  print("common year")
```

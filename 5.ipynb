```python
#4-1  100이하의 양의 정수만 입력된다.
# while 문의 이용하여 1부터 입력받은 정수까지의 합을 구하여 출력하는 프로그램을 작성하시오 ex) 입력 10 / 출력 55

x = int(input())
sum = 0
number = 1
while number <= x:
  sum += number
  number += 1
 
print(sum)
```

```python
#4-2 100이하의 정수를 입력받아서 입력받은 정수부터 100까지의 합을 출력하는 프로그램을 작성하시오.
# ex) 입력 95 / 출력 585

num = int(input())
sum = 0

while num <= 100:
  sum = sum + num
  num = num+1

print(sum)



```

```python
#5-2 2부터 9까지의 수 중 2개를 입력받아 입력받은 수 사이의 구구단을 출력하는 프로그램을 작성하시오. 단 반드시 먼저 입력된 수의 구구단 부터 아래의
# 형식에 맞게 출력해야함 구구단 사이의 공백은 3칸   입력 5 3 / 출력 5단 4단 3단

a, b = map(int, input(). split())

if ((a < 2) or (a > 9)) or ((b < 2) or (b > 9)):
  print()  
elif (a>b):
  for i in range(1, 10):
    for j in range(a, b-1 , -1): 
      print(j , '*' , i , '=' , i*j ,end="  ")
    print()
else : 
  for i in range(1, 10):
    for j in range(a, b+1 , 1): 
      print(i , '*' , j , '=' , i*j , end="  ")      
    print()  
```

```python
#5-3 행과 열의 수를 입력받아 다음과 같이 출력하시오 입력 3 4 / 출력 예 1 2 3 4  ,  2 4 6 8 , 3 6 9 12

a, b = map(int, input(). split())

for i in range(1, a+1):
  for j in range(1, b+1):
    print(i*j, end=" ")
  print()
```

```python
#5-4 자연수 n을 입력받아 n줄 만큼 다음과 같이 출력하는 프로그램 작성   입력 5 / 출력 * ** *** **** *****
n = int(input())

for i in range(1,n+1): #행
  for j in range(1, i+1): #열
    print("*", end="")
  print()
 
```

```python
#5-5 자연수 n을 입력받아서 다음과 같이 출력하시오   입력 3  / 출력 ***** *** *

n = int(input())

for i in range(n):
    for j in range(n,i,-1):
        print(' ',end='')

    for j in range((i+1)*2-1):
        print("*",end='')
    print()
```

```python
#5-8 자연수 n을 입력받아 "출력 예" 와 같이 공백으로 구분하여 출력하는 프로그램을 작성
#주의!! 숫자를 공백으로 구분하되 줄사이에 빈줄을 없다

n = int(input())

number = 1

for i in range(n):
  for j in range(n):
    if i <= j:
      print(number, end=" ")
      number += 1
    else:
      print(' ',end=' ')
  print()
```

```python
#5-9 자연수 n을 입력받아 "출력 예"와 같이 n x n 크기에 공백으로 구분하여 출력 작성
# 10 미만의 홀수만 출력하시오 주의!! 숫자는 공백으로 구분하되 줄 사이에 빈줄은 없다.

n = int(input())

num = 1

for i in range(n):
  for j in range(n):
    print(num, end=' ')
    num += 2
    if num > 10:
      num = 1
  print()
```

```python
#1-1 합계와 평균 수식 내기 kor90, mat80, eng 100, sum 270, avg:
kor =int(input("kor 점수를 입력하시오:"))
mat =int(input("mat 점수를 입력하시오:"))
eng =int(input("eng 점수를 입력하시오:"))

sum = kor + mat + eng
avg = sum//3

print("kor", kor)
print("mat", mat)
print("eng", eng)
print("avg", avg)
```
